1. Line plot the count of "Susan" as the top 25 names over year, where we can see it is decreasing and also see its rate of decreasing.

```python
df_susan = df[df["Name"] == "Susan"]
plt.plot(df_susan["Year"], df_susan["Count"])
plt.title("Count of the Name Susan over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

2. Pie chart showing the Distribution of Top 25 Baby Names by Sex in 1999. Male's popular name numbers are over Female's.

```python
df_1999 = df[df["Year"] == 1999]
sex_counts = df_1999.groupby("Sex")["Count"].sum()

plt.pie(sex_counts, labels=sex_counts.index, autopct="%1.1f%%")
plt.title("Distribution of Top 25 Baby Names by Sex in 1999")
plt.show()
```

3. Popularity of baby names by sex and decade. From the heatmap, we can see the change of popularity and also the count.

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python

df["Decade"] = (df["Year"] // 10) * 10
name_gender_decade_counts = df.groupby(["Decade", "Sex", "Name"])["Count"].sum().reset_index()

name_counts = name_gender_decade_counts.pivot_table(index="Name", columns=["Sex", "Decade"], values="Count", aggfunc="sum")

plt.figure(figsize=(12, 8))
sns.heatmap(name_counts, cmap="coolwarm", cbar_kws={"label": "Count"})
plt.title("Popularity of Baby Names by Gender and Decade")
plt.xlabel("Gender/Decade")
plt.xticks(rotation=45)
plt.ylabel("Name")
plt.tight_layout()
plt.show()

```

4. Bar chart of top 10 names for 2021.

```python

df_2021 = df[df['Year'] == 2021]
df_2021= df_2021.sort_values(by='Count', ascending=False)

top_male_names = df_2021[df_2021['Sex'] == 'Male'].nlargest(10, 'Count')
top_female_names = df_2021[df_2021['Sex'] == 'Female'].nlargest(10, 'Count')

# Create the bar chart
fig, ax = plt.subplots()
ax.bar(top_male_names['Name'], top_male_names['Count'], label='Male')
ax.bar(top_female_names['Name'], top_female_names['Count'], label='Female')
ax.legend()
ax.set_xlabel('Name')
ax.set_ylabel('Count')
plt.xticks(rotation=60)
plt.show()
```

5. Stacked bar chart of top 5 names over time.

```python
top_names = df.groupby(["Sex", "Name"]).sum().sort_values(by="Count", ascending=False).groupby("Sex").head(5).reset_index()["Name"].tolist()
df_top_names = df[df["Name"].isin(top_names) ]
df_top_names = df_top_names.groupby(["Year", "Name", "Sex"]).sum().reset_index()
df_top_names = df_top_names.pivot_table(values="Count", index="Year", columns=["Name", "Sex"], fill_value=0)
df_top_names.plot(kind="bar", stacked=True)
plt.title("Top 5 Names over Time")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
